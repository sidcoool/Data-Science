```python
import bs4
import html5lib
import requests
```

```python
product=input()
```

```python
url="https://www.amazon.in/s?k="
url+=product
print(url)
```

```python
r=requests.get(url)
print(r.content)
```

```python
soup=bs4.BeautifulSoup(r.content)
```

```python
classhtml=soup.findAll('div',{'class':'a-section aok-relative s-image-fixed-height'})

print(classhtml[1])
```

```python
first_class=classhtml[0]
first_class.img.attrs['src']
imgurl=first_class.img.attrs['src']
print(imgurl)
```

```python
first_class=classhtml[0]
first_class.img.attrs['src']
imgurl=first_class.img.attrs['src']
imgresponse=requests.get(imgurl)
with open("amazonimages.jpg","wb") as f:
        f.write(imgresponse.content)
```

```python
for idx,allclass in enumerate(classhtml):
    with open("amazon_{}_images{}.jpg".format(product,idx),"wb") as f:
        imgurl=allclass.img.attrs['src']
        imgresponse=requests.get(imgurl)
        f.write(imgresponse.content)
        
```
