```python
import matplotlib.pyplot as plt
import numpy as np
plt.style.use("seaborn")
```

```python
x=np.array([1,2,3,4,5])
y=x**2
y2=x
plt.plot(x,y,color='red',label="year 2010",linestyle="dashed",marker='o')
plt.plot(x,y2,color="green",label="year 2020")
plt.xlabel("time")
plt.ylabel("money")
plt.title("stock market price over the years")
plt.legend()
plt.show()
```

```python
plt.scatter(x,y)
plt.figure(figsize=(2,2))
plt.show()
```

## bar graph

```python
plt.bar([0,1,2],[10,20,50])
plt.show()
```

```python
xcord=np.array([1,2,3])*2
plt.bar(xcord,[10,25,30],width=0.5,label="current year",tick_label=["gold","silver","platinum"])
plt.bar(xcord+0.5,[15,40,50],width=0.5,label="legend year",color="yellow")
plt.title("metal price comparison")
plt.xlabel("metal")
plt.ylabel("price")
plt.legend()
plt.show()
```

 ## pie chart

```python
subjects=["maths","phsics","chemistry","english"]
weights=[10,20,15,30]
plt.pie(weights,labels=subjects,explode=(0,0,0.11,0))
plt.show()
```

## histogram

```python
xsn=np.random.randn(100)
#print(xsn)
sigma=5
u=60
x=np.round(xsn*sigma+u)
print(x)

x2=np.round(xsn*3+80)
print(x2)
```

```python
plt.style.use('seaborn')
plt.hist(x,label="physics")
plt.hist(x2,alpha=0.7,label="maths")
plt.xlabel("marks of students")
plt.ylabel("number of students")
plt.show()
```

## multivariate normal distribution

```python
# bivariate 
mean=np.array([0.0,0.0])
cov=np.array([[1,0],[0,1]])

mean2=np.array([10.0,30.0])
cov2=np.array([[1.8,0.8],[0.8,0]])

distribution=np.random.multivariate_normal(mean,cov,500)
distribution2=np.random.multivariate_normal(mean2,cov2,500)
#print(distribution)

plt.scatter(distribution[:,0],distribution[:,1] )
plt.scatter(distribution2[:,0],distribution[:,1] )
plt.show()
```
