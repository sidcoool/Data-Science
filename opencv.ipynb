```python
import cv2
```

```python
img=cv2.imread('facebook_dp.jpg')
```

```python
cv2.imshow('dp',img)
cv2.waitKey(0)#wait for infinite time
cv2.destroyAllWindows()
print(img.shape)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.imshow(img)
```

```python
rbg_img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.axis('off')
plt.imshow(rbg_img)
```

```python
crop_img=rbg_img[0:75,50:150,:]
plt.axis('off')
plt.imshow(crop_img)
```

```python
resize_img=cv2.resize(rbg_img,(100,100))
plt.imshow(resize_img)
```
